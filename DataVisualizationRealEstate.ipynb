## Data Visualization - A substantive yet humiliating journey

Think of the thousands of graphs that you have come across in your education, on the web, or in your professional work. Take a second to consider one graph that you came across that instinctually you disliked, maybe not for what information was being communicated - but *how* that information was being communicated. Data visualization is an essential tool to any workflow. It's also an influential part of the entire process of cleaning, wrangling, analyzing, and presenting your data. When we present our ideas, we're attempting to change views, provide perspective, or summarize information succinctly. Whether you are preparing informative graphics for your next team meeting in-house or your major presentation reaching thousands, we can approach the process of data visualization the same way for both scenarios. This requires a basic appreciation for what data visualization should capture and how we can use to convey insights. Follow these guidelines for best practices:

- __Comparison__
    - Convincing others of your findings requires a moderate to an advanced comprehension of the material. To prove the validity of your research, a comparison group could be quite useful. Comparing different groups is optimal using data visualization tools, such as a grouped barplot, as seeing the raw figures themselves in a spreadsheet alone only goes so far. Without a comparison of some kind, if a comparison is appropriate, your research might have very little justification in what you are trying to argue and little influence over convincing others. For example, if we are trying to convince others of a certain trend happening in home sales, we could compare different timeframes if certain influential legislation were passed that impacted this; we also could compare these trends across different geographies for additional context; or we can do both as well.

- __Casuality__
    - Almost always, you try to explain the reason behind certain trends and phenomena. This requires research on the subject matter to appreciate and understand underlying cause-and-effect relationships. It also requires access to the data containing those indicators that provide explanation. Using a causal framework to demonstrate the relationship between a cause and an effect can not only help you refine your research, but can help you master expertise of the topic. Think of the next time you give a presentation and follow-up with a Q&A. One question might relate to how you can further justify your findings or what other indicators that you explored also have influence on the relationship. While graphics themselves should stand alone with little to no explantation for what data is being visualized, sometimes it's that two-minute response to a question that proves how well you are able to elevate the findings in your graph.
    
- __Multivariate__
    - The reason I emphasize needing to have somewhat of a mastery over your data and research is because if things were so simple, we would not need much explanation or convincing - to some extent. Many indicators in the real world are messy and interact with one another; but let's recall that adage that correlation does not prove causation. When we compare only two indicators, we limit our scope and appreciation of the byzantine world in which we live. Think back to the beginning of the pandemic last year when everything was starting to shut down. In Seattle, for example, we saw that as businesses started to reopen, their sales also increased. While this was happening, people's use of public transit also increased. Now, if we were to only consider those two indicators, visualize them to demonstrate their positive relationship, we would be making an uninformed conclusion based on little evidence apart from the fact that both experienced growth around the same time. If we were to take a step back, what else could be driving both trends? Is it all businesses experiencing growth or certain industries? Are people using transit to go to those businesses, or are they going somewhere else?

## Data Visualization Checklist

When starting off your data visualization journey, run through this checklist to ensure total competency and make for a smooth data wrangling and visualization stage in your workflow.

An example would be if you have a set of indicators in one column, and each subsequent, individual column represents a single year. If your dataframe is in long-form, it has one row per observation. So, now, instead of each year having its own column, we have only one column denoting the year, and the different levels or categories of that indicator variable repeating for every single year it is recorded.

### Data Viz - After

```python
# create new date column
statenisland_df["sales_year"] = pd.to_datetime(statenisland_df["sales_year"])

# create new column dividing sales price by square footage
statenisland_df['priceunit'] = statenisland_df['price']/statenisland_df['unit_size']

# create new column dividing sales price by square footage
statenisland_df['priceland'] = statenisland_df['price']/statenisland_df['land_size']

# create new column dividing sales price by square footage
statenisland_df['pricesqft'] = statenisland_df['price']/statenisland_df['unit_size']

# create dataframe calculating median sales price/square footage by year
grouped_price = statenisland_df.groupby(['sales_year','todt'], as_index=False).agg({
         'pricesqft': "median",
})

grouped_price.head()
```

## Debugging & Demystifying Data Visualization

### Debugging 

Introducing exploratory visualizations from the outset to the presentation of your research helps you identify potential errors and mistakes in the data. Examining your data visually will save you time by troubleshooting those problems in the earlier stages.

* For text, this means to check for minor spelling errors, capitalization, or extra blank spaces (e.g., Agree  and agreee) by creating count plots for categorical variables to see all possible levels.
* For numeric figures, this means to check for formatting (e.g., percentage, currency), missing values, or outliers, or even missing values coded as outliers (e.g, 999999).
* For dates, this means to standardize the format (e.g., YYYY-MM-DD), and to provide alternative variables that reflect only the year, month, or both for summarizing the data down the road.

### Demystifying

Creating convincing information visualizations need not be overwhelming. The temptation to competitively employ an intricate 3D graphic or the most recent open-source tool yet may cause unnecessary anxiety over presenting your research. It is best to create those animated and intricate information visualizations specifically for their own presentation or platform (e.g., dashboard). If you are presenting your work to a live audience especially, stick to the foundational graph formats (e.g., point plot, grouped/stacked bar plot, donut plot) and aesthetics to avoid confusion and misinterpretation.

### Before

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
sns.scatterplot(
    data = statenisland_df,
    y = 'priceunit', x = 'priceland', 
    hue = "todt", hue_order=[1,0],
    palette=color_order)
plt.tight_layout()
plt.show()
```

### After

```python
# create summary dataframe
summary_price = statenisland_df.groupby(['sales_year','todt'], as_index=False).agg({
    'pricesqft': "median",
    'priceunit': "median",
    'priceland': "median",
})

# create new date column
summary_price["sales_year"] = pd.to_datetime(summary_price["sales_year"])

summary_price.head()
```

```python
# make the pairgrid as baseline
g = sns.PairGrid(summary_price.sort_values("pricesqft", ascending=False),
                 x_vars=summary_price.columns[2:], y_vars=["sales_year"],
                 hue='todt',  hue_order=[1,0],
                 palette=color_order,
                 height=8, aspect=.50)

# create dot plot
g.map(sns.stripplot, size=10, orient="h", jitter=False)

g.set(xlabel="Price", ylabel="")

# rename columns
titles = ["Price/Sqft", "Price/Unit", "Price/Land"]

for ax, title in zip(g.axes.flat, titles):

    # set a different title for each axes
    ax.set(title=title)

    # make the grid horizontal instead of vertical
    ax.xaxis.grid(True)
    ax.yaxis.grid(True)

sns.despine(left=False, bottom=False)
```

## Connecting & Communicating Visually

Now, in both your exploratory and explanatory phases, keep the overall design concept of the graph in mind. What type of data visualization do you envision using? Why does that work best? What would be an alternative based on your variable types and number of variables? In order to understand your insight that you spent all that time uncovering, figure out what kinds of trends or continuity the audience should see. For example, if a particular figure youre interested in fluctuates overtime, the audience should be able to appreciate that continuity as well. Also, consider what kind of contrasts the audience should see. For example, what would be a comparable trend to showcase against the current trend were highlighting? What would be another group, city, or country that could provide additional context? How far back in the data should we go, or what timeframe should we employ, in order to denote abrupt or slow fluctuations?

Design can help you strategize your data visualization effectively and productively. The process of data visualization is a series of decisions you make as far as what variables you will use, and where and when you will use them in the graph. Consider for each relevant variable, what aesthetics you might use - what should go on your Y-axis, your X-axis, what colors should you use, what kind of color scale - discrete, continuous, or gradient - line type, size, font, and so forth. Be sure the decisions you make are appropriate for making the continuity and contrasts you want to communicate. This might feel either a given or a tad overwhelming, so think of design in terms of layers. We start off with the base layer, that is, the outline or template of the type of graph we will use. Theres a y and x axis. But nothing has been assigned yet. If were visualizing percentage share with a stacked bar plot, would it be better to have the bars running vertically or horizontally? Meaning, in our next layer, should we think of our y axis as continuous or discrete? After making that decision, we slowly build on each aspect of the graph until we have thought through each part and can justify every single decision we made in constructing the final graph. 

### Connecting

"Connecting": Connect with an audience by visually orienting them through compelling graphics or visualizations that help translate what you most want to convey about your research. Because you are not visualizing raw data, you already have made assumptions and decisions about the structure of your data in preparation for your intended analysis. What makes graphics and visualizations compelling, similar to wrangling the structure of the data itself, rests on your point of view. Analysis can inform and build intuition - but without thoughtfully-executed judgement, analysis alone can not assess context fully or make decisions holistically (Carroll 2003).

* It is important to define compelling as having the ability to simply and efficiently inform and persuade your audience. Compelling graphics and visualizations need not be complex and over-ridden with facts and design features. Having too much in one space can hinder your audiences ability to distinguish the message from the noise. While this seems more abstract and less concrete an idea, consider what graphics and visualizations have persuaded or influenced you before and why.

### Communicating

"Communicating": Every aesthetic and feature in each of your information visuals communicates something about your research to an audience. The level of detail shown can depict your level of sophistication and professionalism, as well as your zeal and creativity. Minute details - from including a thousands separator (10,000 versus 10000) to date format (May 2020 versus 2020-05) - communicate your dedication, meticulousness, and expertise in extraordinary ways.

* Similar to how Jacques Bertin, a French cartographer, frames it, you can think of representing data utilizing geometric objects: a point, line, an area, or a geographic map. To display the differences within each representation (i.e., group, category, time, range), change the variables in terms of size, shape, color, value (e.g., opacity), texture, and orientation.
* Although you could argue over how perfect and representative you are, your data and model almost never will be. Depending on your field and your audience, representing statistical error is a given. Displaying this error adds to your narrative, but it should not visually distract or detract attention. Applying a light gray color palette with a solid line or shading is a sound approach.

"Aesthetics": After choosing a type of graphic or visualization, and establishing a title, subtitle, legend format, and axis labels (though, depending on how clear your title/subtitle and legend are, you may not need labels), think of the remaining aesthetics choices before you as a point system. This tutorial on San Francisco home prices does a thorough job of explaining how to apply solid aesthetics in spatial visualizations.

* __Point System__: Would you gain or lose points if you added additional features (e.g., color, shapes, themes, groups, time-series) to the plot? Would you lose points if you removed certain elements from the plot? Consider how doing either could be alternatively executed. 
* For example, say you decide to limit yourself choosing only a select number of aesthetic features within a time-series chart you are creating. You will lose points if you include all pairs of groups that you could possibly compare, but you will gain points if you decide to only compare the top or bottom three. You will gain a point if you use a discrete color palette (one color per group), but you will lose a point if you cannot connect the meaning of the assigned colors to each group. If you are confused about whether a certain aesthetic choice will increase or decrease your points, ask yourself whether making that choice will enhance your ability to simply and efficiently inform and persuade your audience.

## Audience: Game Time

### Presentation

You conducted your research, constructed an appropriate deliverable, and produced the graphics and visualizations you deem necessary. Now the audience members will test how well you can defend your work. When interacting with an audience in-person or online, remember that they will be as critical of your information visuals as you are of your modelling.

* __Justification__: The aesthetics and placement of each graphic or visualization in your deliverable rests entirely on your judgement. When your audience rightfully inquires about your choices, you will need to provide sound justification explaining every detail you include - and likely did not include. Providing justification for these details requires you to express an endearing kind of thoughtfulness that both humanizes you and demonstrates a robust evaluation process that suggests objectivity and prudence. Be prepared to defend your choices, both in terms of your data analysis and how that was presented. If something in your analysis or presentation makes you uneasy, it will likely make your audience uneasy, too.
* __Humility__: Your engagement with an audience is an equal opportunity for you both to learn from one another. Demonstrate humility and collaboration by applying the audiences critical input towards improving your present and future work. If perfecting your present work means removing a graphic that confused the majority of your audience but that you spent 5 hours creating, learn the lesson now to save yourself from wasting 5 more hours on something similar in the future.
* __Transparency__: Avoid misleading graph tactics. This ranges from distorting the vertical or horizontal axis to obscure a comparison or trend, such as starting the vertical axis at a figure other than zero or aggregating the data in a particular way over time (e.g., rolling averages, skipped years, etc.) that overlooks the impact of a particular event. This also includes establishing an interpretable color scheme that does not confuse the audience, such as using lighter colors to make it harder to see or unflattering colors to portray unfavorable points. Moreover, be careful when filtering or subsetting data, as this can also have an impact on what data you visualize. "Cherry-picking" certain data points to falsely depict an unreal phenomena for your graphic could lead to serious consequences if another person were to find discrepancies when analyzing and visualizing the data.

## Quick Tips & Tricks

Graphic Choice -
* __Selection__: Use https://www.data-to-viz.com/ for selecting a data visualization based on your variables. Code is provided for creating each graphic.

Colors -
* __Discrete variables__: Use a discrete color scheme with not too high but not too low contrast. Learn to create custom palettes at Data Novia (https://www.datanovia.com/en/blog/ggplot-colors-best-tricks-you-will-love/).
* __Continuous variables__: The most effective color scheme for a continuous variable is a discrete color scheme that has a gradient feel by including colors that naturally blend together (e.g., blue/purple/red; blue/green/yellow). See the viridis color palettes for inspiration (https://cran.r-project.org/web/packages/viridis/vignettes/intro-to-viridis.html). Avoid color palettes that have color ranges that are uneven and non-uniform, like the rainbow palette (e.g., the rainbow palette has a very short range for yellow but a very large range for green).Note: Only use a single-hue palette (e.g., light red to dark red) if you want to emphasize the difference between the highest and lowest possible values. Any figures in between will get lost.

Values -
* __Numeric values__: Use a thousands-separator and up to two to three decimal points. Include a percentage sign if a percentage, or a currency symbol if a monetary value.
* __Categorical values__: If ordering different groups, rank the groups with a logical hierarchy. Order the categories alphabetically, sequentially, or by value.
* __Dates__: Use a combination of year, month, and/or day depending on timeframe and axis space. Date variables should be assigned to the x-axis almost always.

Legend -
* __Increments__: Use natural increments on your axes (0, 5, 10, 15, 20) instead of awkward or uneven increments (0, 3, 5, 16, 50). See ColumnFiveMedia for additional insight.
